```python
import numpy as np
import pandas as pd
import os
import cv2
from tqdm import tqdm

img_dir = "processed_images"
image_size = 224
```

```python
df = pd.read_csv("data.csv")
df.reset_index()

processed_df = pd.DataFrame(columns=["file", "x", "y"])

df
```

```python
for i in tqdm(range(df.shape[0])):
    data = df.iloc[i]
    image = cv2.imread(os.path.join(img_dir, data['file']))
    height, width, _ = image.shape
    
    processed_y = int((data['y']*image_size) / height)
    processed_x = int((data['x']*image_size) / width)
    
    processed_data = {'file' : data['file'], 'x': processed_x, 'y': processed_y}
    processed_df = processed_df.append(processed_data, ignore_index=True)
```

```python
processed_df
```

```python
processed_df.to_csv("processed_data.csv")
```

```python
train = processed_df.iloc[:125*8]
valid = processed_df.iloc[125*8:125*9]
test = processed_df.iloc[125*9:]
```

```python
train.set_index("file", inplace=True)
valid.set_index("file", inplace=True)
test.set_index("file", inplace=True)



print(train.shape)
print(valid.shape)
print(test.shape)
```

```python
train.to_csv("train.csv")
valid.to_csv("valid.csv")
test.to_csv("test.csv")
```
